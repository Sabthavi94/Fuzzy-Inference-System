Step 2 - Generate Membership Functions for Input and Output Variables

```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
age = ctrl.Antecedent(np.arange(18, 81, 1), 'age')

age['young'] = fuzz.trimf(age.universe, [18, 26, 34])
age['mid_age'] = fuzz.trimf(age.universe, [33, 44, 55])
age['old'] = fuzz.trimf(age.universe, [54, 67, 80])

age.view()
```

```python
bmi = ctrl.Antecedent(np.arange(0, 51, 1), 'bmi')

bmi['underweight'] = fuzz.trimf(bmi.universe, [0, 9, 18.5])
bmi['normal'] = fuzz.trimf(bmi.universe, [18.4, 21 , 24.9])
bmi['overweight'] = fuzz.trimf(bmi.universe, [24.8,27 , 29.9])
bmi['obesity'] = fuzz.trimf(bmi.universe, [29.8, 39, 50])

bmi.view()
```

```python
polyuria = ctrl.Antecedent(np.arange(0, 16, 1), 'polyuria')

polyuria['low'] = fuzz.trimf(polyuria.universe, [0, 2, 4])
polyuria['high'] = fuzz.trimf(polyuria.universe, [3, 6, 9])
polyuria['veryhigh'] = fuzz.trimf(polyuria.universe, [8, 11, 15])

polyuria.view()
```

```python
nocturia = ctrl.Antecedent(np.arange(0, 9, 1), 'nocturia')

nocturia['low'] = fuzz.trimf(nocturia.universe, [0, 1, 2])
nocturia['high'] = fuzz.trimf(nocturia.universe, [1, 2, 4])
nocturia['veryhigh'] = fuzz.trimf(nocturia.universe, [3, 5, 8])

nocturia.view()
```

```python
polydipsia = ctrl.Antecedent(np.arange(0, 15, 1), 'polydipsia')

polydipsia['low'] = fuzz.trimf(polydipsia.universe, [0, 2, 4])
polydipsia['high'] = fuzz.trimf(polydipsia.universe, [3, 5, 7])
polydipsia['veryhigh'] = fuzz.trimf(polydipsia.universe, [6, 10, 14])

polydipsia.view()
```

```python
polyphagia = ctrl.Antecedent(np.arange(1, 11, 1), 'polyphagia')

polyphagia['low'] = fuzz.trimf(polyphagia.universe, [1, 2, 3])
polyphagia['high'] = fuzz.trimf(polyphagia.universe, [2, 4, 5])
polyphagia['veryhigh'] = fuzz.trimf(polyphagia.universe, [4, 7, 10])

polyphagia.view()
```

```python
output = ctrl.Consequent(np.arange(4, 31, 1), 'BloodSugarLevel-mmol/L')

output['nondiabetes'] = fuzz.trimf(output.universe, [4, 7, 9])
output['diabetes'] = fuzz.trimf(output.universe, [7.8, 17, 30])

output.view()
```

Step 4 - Rule Generation

Step 4 - Fuzzy Rules declaration

```python
rule1 = ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['obesity'] & polyphagia['high'] & nocturia['veryhigh'], output['diabetes'])
rule2 = ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['obesity'] & polyphagia['high'] & nocturia['low'] & polyuria['low'] , output['diabetes'])
rule3 = ctrl.Rule(age['old'] & bmi['normal'] & nocturia['low'], output['nondiabetes'])
rule4 = ctrl.Rule(age['old'] & nocturia['veryhigh'], output['diabetes'])
rule5 = ctrl.Rule(age['mid_age'] & polydipsia['low'] & polyuria['high'], output['diabetes'])
rule6 = ctrl.Rule(age['old'] & nocturia['low'] & bmi['overweight'] & polydipsia['veryhigh']  & polyphagia['high'] , output['diabetes'])
rule7 = ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['normal'] & nocturia ['low'], output['nondiabetes'])
rule8 = ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['overweight'], output['diabetes'])
rule9 = ctrl.Rule(age['old'] & nocturia['low'] & bmi['overweight'] & polydipsia['veryhigh'] & polyphagia['high'] & polyuria['low'] , output['diabetes'])
rule10= ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['normal'] & nocturia['veryhigh'], output['diabetes'])
rule11= ctrl.Rule(age['mid_age'] & polydipsia['low'] & polyuria['low'], output['nondiabetes'])
rule12= ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['obesity'] & polyphagia['high'] & nocturia['high'], output['diabetes'])
rule13= ctrl.Rule(age['old'] & nocturia['low'] & bmi['overweight'] & polydipsia['low'] , output['diabetes'])
rule14= ctrl.Rule(age['old'] & nocturia['low'] & bmi['obesity'] , output['diabetes'])
rule15= ctrl.Rule(age['young'] , output['nondiabetes'])
rule16= ctrl.Rule(age['old'] & nocturia['high'] & polyuria['veryhigh']  , output['diabetes'])
rule17= ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['underweight'] , output['diabetes'])
rule18= ctrl.Rule(age['mid_age'] & polydipsia['high'] , output['nondiabetes'])
rule19= ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['obesity'] & polyphagia['veryhigh'] , output['diabetes'])
rule20= ctrl.Rule(age['old'] & nocturia['low'] & bmi['underweight'] , output['diabetes'])
rule21= ctrl.Rule(age['old'] & nocturia['high'] & polyuria['low'] , output['nondiabetes'])
rule22= ctrl.Rule(age['mid_age'] & polydipsia['veryhigh'] & bmi['obesity']  & nocturia['low']  & polyphagia['high'] & polyuria['high'], output['diabetes'])
rule23= ctrl.Rule(age['old'] & nocturia['high'] & polyuria['high'], output['diabetes'])
rule24= ctrl.Rule(age['old'] & nocturia['low']  & bmi['overweight'] & polydipsia['veryhigh'] & polyphagia['high'] & polyuria['veryhigh'], output['diabetes'])
rule25= ctrl.Rule(age['mid_age'] & polydipsia['veryhigh']  & bmi['normal'] & nocturia['high'] , output['diabetes'])
rule26= ctrl.Rule(age['old'] & nocturia['low']  & bmi['overweight'] & polydipsia['veryhigh'] & polyphagia['veryhigh'], output['diabetes'])
rule27= ctrl.Rule(age['mid_age'] & polydipsia['low'] & polyuria['veryhigh'], output['nondiabetes'])
```

Step5 - Fuzzy Inference System Creation & Simulation

```python
fuzzy_ctrl = ctrl.ControlSystem([rule1, rule2,rule3,rule4,rule5,rule6, rule7,rule8,rule9,rule10, rule11, rule12,rule13,rule14,rule15,rule16, rule17,rule18,rule19,rule20,rule21, rule22,rule23,rule24,rule25,rule26, rule27])
rules = ctrl.ControlSystemSimulation(fuzzy_ctrl)
```

```python
# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
rules.input['age'] = 69
rules.input['bmi'] = 30
rules.input['polyuria'] = 6
rules.input['nocturia'] = 2
rules.input['polydipsia'] = 5
rules.input['polyphagia'] = 3

rules.compute()

print(rules.output['BloodSugarLevel-mmol/L'])
output.view(sim=rules)
```

```python
# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
rules.input['age'] = 40
rules.input['bmi'] = 24
rules.input['polyuria'] = 4
rules.input['nocturia'] = 0
rules.input['polydipsia'] = 5
rules.input['polyphagia'] = 3

rules.compute()

print(rules.output['BloodSugarLevel-mmol/L'])
output.view(sim=rules)
```

```python
print("-------------------------------")
age = input("Enter your Age ")
weight= float(input("Enter your weight(kg) "))
height= float(input("Enter your height(m) "))
polyuria= input("How often you urinate during day time ")
nocturia= input("How often you urinate in Night ")
polydipsia= input("How often you feel thirsty per Day ")
polyphagia= input("How often you feel hungry per Day ")
print("-------------------------------\n\n")

bmi = weight/(height*height)

rules.input['age'] = int(age)
rules.input['bmi'] = int(bmi)
rules.input['polyuria'] = int(polyuria)
rules.input['nocturia'] = int(nocturia)
rules.input['polydipsia'] = int(polydipsia)
rules.input['polyphagia'] = int(polyphagia)

rules.compute()

# print(rules.output['BloodSugarLevel-mmol/L'])
# output.view(sim=rules)

def fuzzify_Output(output):
    Mf_nondiabetes =  np.array([4, 7, 11])
    Mf_diabetes =  np.array([7.8, 17, 30])
   
    Array = np.array([output])

    set1=fuzz.membership.trimf(Array,Mf_nondiabetes)
    set2=fuzz.membership.trimf(Array,Mf_diabetes)
    
    d = {'Non_Diabetes':set1, 'Diabetes':set2}

    return max(d, key=d.get)
                                
if(fuzzify_Output(rules.output['BloodSugarLevel-mmol/L'])== 'Diabetes'):
                                print("You have Diabetes")
else:
                                print("You don't have Diabetes")
```

Step 6 - Fuzzy Inference System Evaluation

```python
import csv
with open('testData.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
```

```python
actual_results=[]
predicted_results=[]

for x in data:
        rules.input['age'] = int(x[2])
        rules.input['bmi'] = int(x[5])
        rules.input['polyuria'] = int(x[7])
        rules.input['nocturia'] = int(x[8])
        rules.input['polydipsia'] = int(x[9])
        rules.input['polyphagia'] = int(x[10])
        rules.compute()
        result = fuzzify_Output(rules.output['BloodSugarLevel-mmol/L'])
        predicted_results.append(result)
        actual_results.append(x[15])

print('Actual Results: ', actual_results)
print('Predicted Results: ', predicted_results)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
results = confusion_matrix(actual_results, predicted_results)
print(results)
```

```python
accuracy_score(actual_results, predicted_results)
```
